Linear regression is commonly used in predecting house price , sales , salaries , stock prices , and other numerical values .

Q4. Explain the following evaluation metrics used in Linear Regression:

1.SSE

2.MSE

3.RMSE

4.R² Score

5.Adjusted R²

ans- 1.SSE (sum of squared Errors ) - SSE is the total of the sqaured differences between the actual values and the predicted values . it meausres how well the regression model fits the data . A lOWER SSE indicates a better model because the prediction errors are smaller .
 formula
 [ SSE = \sum (yᵢ - ŷᵢ)^2 ]
 where
 . yᵢ = Actual value
 . ŷᵢ = Predicted value

 2. MSE(Mean Squared Error) - MSE is the average of the squared prediction errors. it is obtained by dividing SSE by the total number of observations .  A LOWER MSE indicates better prediction accuracy .

 formula
 Mean Squared Error = (1/n) * Σ(Yi - Ŷi)^2
where:
. n = is the number of observations.
. Yi =is the actual value of the observation.
.Ŷi= is the predicted value from the regression model.

3. RMSE (Root Mean Squade Error )- RMSE is the square root of MSE . it measures the average prediction error in the same units as the target variable , mking it easier to interper . A smaller RMSE  indicates a more accurate model .

formula

$\mathrm{RMSE} = \sqrt{mathrm{MSE}}$


4.  R²score - the R² score measures how well the regression model expains the variation in the dependent variable . its value ranges from 0 to 1 , An R² value closer to 1 means the model explains the data very well .

formula

 R² = 1 - SSE\SST
where
SSE = sum of squared Errors
SST = total sum of squares

5. Adjust R² - Adjust R² is an improved version of R² that considers the number of independent variables oin the  model . it penalizes the addition of unnecessary variable , making it more relaiable for evaluating multiple linear regression models .

formula
 adjusted R² = 1- (1-R²) (n-1/n-p-1)

 where
 . n= Number of observations
 . p = Number of independent variables

Q8. Select independent variables (features) and dependent variable (target).

```python
from os import XATTR_CREATE
# select features (X) and target (Y)

X = df.drop('medv', axis=1)
# independent variables
Y = df['medv']
# dependent variable (Target)

print("Features (X):")
print(X.head())

print("\nTarget (Y):")
print(Y.head())
```

Q9. Split the dataset into training and testing sets (80% training, 20% testing).

```python
from sklearn.model_selection import train_test_split

# split the dataset into training (80%) and testing (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# display the shape of the dataset
print("X_train shape:", X_train.shape)
print("Y-train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

Q10. Train a Linear Regression model and perform the following:

Make predictions

Calculate MAE

Calculate MSE

Calculate RMSE

Calculate R² Score

Calculate Adjusted R²

Write the Linear Regression equation

Interpret the model coefficients

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
import numpy as np

# Train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Make predictions
Y_pred = model.predict(X_test)

# calculate evaluation metrics
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_pred)

# calculate adjusted R²
n = len(Y_test)
p = X_test.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/ (n-p-1)

# print results
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("r2 Score:", r2)
print("Adjusted R²:", adjusted_r2)

# Linear Regression Equation
print("\nIntercept:", model.intercept_)
print("Coefficients:")
for feature , coef in zip(X.columns,
model.coef_):
     print(feature, ":", coef)
```
